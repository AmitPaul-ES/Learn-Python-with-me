```python
n = [3,6,2,8,9,56]
total = 0
for x in n:
    total = total + x
    print(total)
print('Total',total)
```

```python
for i in range(10): # loop always run to 'n-1'
    print("Data science",i)
```

```python
for i in range(5,10): # loop always run to 'n-1'
    print("Data science",i)
```

```python
for i in range(100,50,-10): # loop always run to 'n-1'
    print("Data science",i)
```

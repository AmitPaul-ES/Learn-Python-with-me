```python
total_sum = 0
for num in range(10, 1001):
    if num > 1: 
        is_prime = True
        for i in range(2, int(num**0.5)+1):
            if num%i == 0:
                is_prime = False
                break
        if is_prime:
            total_sum += num
print(total_sum)
```

```python
n = 7
factorial = 1
for i in range(1, n+1):
    factorial = factorial*i
print("The factorial of n is:",factorial)
```
